```python
import matplotlib.pyplot as plt
date=['22/12','26/12','27/12']
temp=[8.5,10.5,6.8]
plt.plot(date,temp,label='temp')
plt.xlabel("Date")
plt.ylabel("temperature")
plt.title("Date wise temperature")
plt.grid(True)
plt.yticks(temp)
plt.legend(("Blue","Green"))
plt.show()
plt.savefig("x.png")
```
